### Import Libraries

```python
# Import NumPy and Pandas
import pandas as pd
```

```python
# Read the downloaded dataset
data = pd.read_csv(path + "/valentine_gifts_satisfaction.csv")

data.head()
```

### Data Explanation

| Column | Type | Description |
|--------|------|-------------|
| `id` | Integer | Unique identifier (1-670) |
| `receiver_gender` | String | Gender of gift receiver (Female, Male, Non-binary) |
| `receiver_age` | Integer | Age of receiver (18-60 years) |
| `relationship_type` | String | Type of relationship (Dating, Married, First Date, Engaged, Complicated) |
| `months_together` | Integer | Duration of relationship in months (0-357) |
| `gift_item` | String | Specific gift given (18 different items) |
| `gift_category` | String | Category of gift (Sweets, Jewelry, Luxury, Collectibles, Gaming, Sentimental) |
| `price` | Float | Price of gift in USD ($5-$500) |
| `flower_count` | Integer | Number of flowers included (0, 1, 3, 6, 12, 24, 50, 67) |
| `days_prepared` | Integer | Days spent preparing the gift (0-200) |
| `receiver_satisfaction` | Integer | Satisfaction rating (1-5 scale, 1=very unhappy, 5=very happy) |
