## Pandas Data Structures

There are 2 main data structures in Pandas:
- **Series**: 1D homogenous (single typed) array 
- **DataFrame**: 2D tabular structure with potentially heterogenous (different typed) columns

### DataFrame

we can think of a DataFrame as a bunch of Series objects put together to share the same index


```python
employees = {
    'name': ['Alice Chen', 'Ben Sok', 'Clara Lim', 'Daniel Wong', 'Emily Tran'],
    'gender': ['F', 'M', 'F', 'M', 'F'],
    'job': ['Data Analyst', 'Project Manager', 'Software Engineer', 'Data Scientist', 'UI/UX Designer'] 
}
```

```python
employees_df = pd.DataFrame(employees)
```

```python
employees_df
```

```python
employees_ages = [30, 28, 42, 24, 23]
```

```python
employees_df['age'] = employees_ages
```

```python
employees_df
```

```python
education_data = {
    'university': ['AUPP', 'PIU', 'ITC', 'RUPP', 'AUPP'],
    'gpa': [3.66, 3.12, 2.12, 3.84, 3.12]
}
```

```python
education_df = pd.DataFrame(education_data)
```

```python
employees_df = pd.concat([employees_df, education_df], axis=1)
```

```python
employees_df
```

### Descriptive Statistics

```python
# Statistics on age column
print("Age Statistics:")
print(f"  Count: {employees_df['age'].count()}")
print(f"  Mean: {employees_df['age'].mean()}")
print(f"  Median: {employees_df['age'].median()}")
print(f"  Mode: {employees_df['age'].mode().values}")
print(f"  Std Dev: {employees_df['age'].std()}")
print(f"  Min: {employees_df['age'].min()}")
print(f"  Max: {employees_df['age'].max()}")
```

```python
age_gpa_corr = employees_df['gpa'].corr(employees_df['age'])

age_gpa_corr
```

```python
import matplotlib as plt
import seaborn as sns

sns.regplot(data=employees_df, x='age', y='gpa')
```
